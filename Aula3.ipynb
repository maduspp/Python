1) Seu programa deve ler palavras e frases, uma por linha, e repeti-la como um loud parrot

```python
lista = []
frase = str(input())

while frase != '.':
  f = frase.title()
  maiuscula = f.upper()
  lista.append(maiuscula)
  frase = str(input())

if frase == '.':
  lista.append('BYE')

print('\n'.join(map(str, lista)))


```

4) Carlos Drummond de Andrade #3

> Seu programa deve ler um número indeterminado de strings, e imprimir o número de strings lidas como saída desde que a string não seja nula, ou seja, vazia. O programa deve terminar quando a string CDA 1942 for lida.



```python
total = 0
verso = input()

while verso != 'CDA 1942':
  if len(verso) == 0:
    total = total + 0
    verso = input()
  else:
    total = total + 1
    verso = input()

print(total)
```

6) Infinito enquanto dure
> Seu programa deve

consumir a sua própria condição de parada e, depois,

consumir linhas da entrada até que seja fornecida a condição de parada lida originariamente, contando o número de linhas consumidas

```python
parada = input()
linha = input()
total = 0

while linha != parada:
  total += 1
  linha = input()

print(total)
```

13) Dispositivos inteligentes

```python
condicao = input()
m = 0
p = 1

while condicao != 'f':
  if condicao == 'm':
    m += 1
    condicao = input()
  elif condicao == 'p':
    p += 1
    condicao = input()
  elif condicao == 'f':
    print('sedentário')

if m > p:
  print('ativo')
else:
  print('sedentário')
```

15) Aceitos recebidos

```python
resposta = input()
certo = 0
errado = 0

while resposta != 'timeout':
  if resposta == 'accepted':
    certo += 1
    resposta = input()
  elif resposta == 'wrong answer':
    errado += 1
    resposta = input()

print(certo)
```
