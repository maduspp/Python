11) Email #1
quantos NÂO são emails validos

```python
numero = int(input())
qn = 0
qe = 1

while qe <= numero:
  email = str(input())
  qe += 1
  if '@' not in email:
    qn += 1

print(qn)
```

12) Email #2

```python
numero = int(input())
qn = 0
qe = 1

while qe <= numero:
  email = str(input())
  qe += 1
  if '@gmail.com' not in email:
    qn += 1

print(qn)
```
